```python
color = (255, 0, 0)
thickness = 2

# Ancient Lives classes
classes = ['Alpha', 'Beta', 'Chi', 'Delta', 'Epsilon', 'Eta', 'Gamma', 'Iota', 
           'Kappa', 'Lambda', 'Mu', 'Nu', 'Omega', 'Omicron', 'Phi', 'Pi', 'Psi',
           'Rho', 'Lunate Sigma', 'Tau', 'Theta', 'Upsilon', 'Xi', 'Zeta']

greek = ['\u0391', '\u0392', '\u03A7', '\u0394', '\u0395', '\u0397', '\u0393', '\u0399', 
         '\u039A', '\u039B', '\u039C', '\u039D', '\u03A9', '\u039F', '\u03A6', '\u03A0', '\u03A8', 
         '\u03A1', 'C', '\u03A4', '\u0398', '\u03A5', '\u039E', '\u0396']

'''goodchars = [[83, 76],
            [18, 34, 26, 86, 31, 8, 5],
            [9, 66, 89, 21, 46, 52, 1, 7, 42, 6, 33, 80, 84, 78],
            [68, 55, 81, 96, 28, 54, 17, 56, 44, 57, 58, 29, 19, 79],
            [49, 3, 61, 30, 4, 59, 10, 47, 60, 15],
            [12, 0, 32, 43, 69, 62, 53, 20, 24, 74, 45],
            [22, 14, 72, 88, 11, 71, 37, 97, 73, 51, 36],
            [63, 2, 39, 35, 82, 48, 41, 65, 16, 64, 77, 50, 38, 100],
            [98, 25, 70, 23, 27, 13, 40, 67]]'''

selchars = [[98, 25, 70, 23, 27, 13, 40, 67],
            [63, 2, 39, 35, 82, 48, 41, 65, 16, 64, 77, 50, 38, 100],
            [22, 14, 72, 88, 11, 71, 37, 97, 73, 51, 36],
            [12, 0, 32, 43, 69, 62, 53, 20, 24, 74, 45],
            [49, 3, 61, 30, 4, 59, 10, 47, 60, 15],
            [68, 55, 81, 96, 28, 54, 17, 56, 44, 57, 58, 29, 19, 79],
            [9, 66, 89, 21, 46, 52, 1, 7, 42, 6, 33, 80, 84, 78],
            [18, 34, 26, 86, 31, 8, 5],
            [83, 76]]

for i in range(len(selchars)):
    selchars[i].reverse()
```

```python
infer = load_model("static/models/alv2_cxe")
```

```python
df = pd.read_csv('static/data/current.csv', index_col=0)
xc = list(df['xc'])
yc = list(df['yc'])
w = list(df['width'])
h = list(df['height'])
```

```python
dlines = []
```

```python
for i in range(len(selchars)):
    line = selchars[i]
    currline = []
    print("Line: ", i)
    for j in range(len(line)):
        cid = line[j]
        xmin = int(xc[cid] - (w[cid] / 2))
        ymin = int(yc[cid] - (h[cid] / 2))
        xmax = int(xc[cid] + (w[cid] / 2))
        ymax = int(yc[cid] + (h[cid] / 2))
        cropped = img.crop((xmin, ymin, xmax, ymax))#
        #cropped = img[xmin:xmax, ymin:ymax]
        cropped = cropped.resize(outsize)
        #x.append(cropped)
        
        X = []
        res = np.array(cropped)
        X.append(res)
        x = np.array(X)
        preds = infer.predict(x)
        imclass = np.argmax(preds)
        char = classes[imclass]
        prob = preds[0][imclass]
        #info = 'Line: ' + str(i) + '   Character:' + str(j) + '   Probability: ' + str(prob) + "   Prediction: " + char
        
        #plt.figure()
        #plt.title(info)
        #plt.imshow(cropped)
        
        currline.append([imclass, prob])
        
    print("\n*********************************************\n")
    dlines.append(currline)
```
